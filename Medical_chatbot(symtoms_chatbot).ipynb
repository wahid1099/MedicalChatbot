```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df_train=pd.read_csv("/content/drive/MyDrive/Software Project 2/symtoms tracker/Training_symtoms.csv")
```

```python
df_test=pd.read_csv("/content/drive/MyDrive/Software Project 2/symtoms tracker/Testing_symtoms.csv")
```

```python
df_train
```

```python
df_test
```

```python
dimensionality_reduction = df_train.groupby(df_train['prognosis']).max()
```

```python
X = df_train.iloc[:, 0:132].values
y = df_train.iloc[:, -1].values
```

```python
dimensionality_reduction
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 5)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
# Displaying the confusion matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python

# Displaying the classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB
model3 = GaussianNB()
model3.fit(X_train, y_train)
y_pred3 = model3.predict(X_test)

from sklearn.metrics import accuracy_score
print("Accuracy Score:",accuracy_score(y_test,y_pred3))
```
